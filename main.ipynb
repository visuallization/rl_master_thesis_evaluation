```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

unity_ml_agents = pd.read_csv("./logs/unity_ml_agents.csv")
godot_rllib = pd.read_csv("./logs/godot_ray_rllib.csv")
godot_sf = pd.read_csv("./logs/godot_sample_factory.csv")
godot_sb3 = pd.read_csv("./logs/godot_stable_baselines_3.csv")
godot_crl = pd.read_csv("./logs/godot_clean_rl.csv")
```

```python
plt.plot(unity_ml_agents.Step, unity_ml_agents.Value, label="Unity ML-Agents")
plt.plot(godot_rllib.Step, godot_rllib.Value, label="Godot RL Agents [RLlib]")
plt.plot(godot_sf.Step, godot_sf.Value, label="Godot RL Agents [Sample Factory]")
plt.plot(godot_crl.Step, godot_crl.Value, label="Godot RL Agents [CleanRL]")
plt.plot(godot_sb3.Step, godot_sb3.Value, label="Godot RL Agents [Stable Baselines 3]")
plt.xlabel("Steps")
plt.ylabel("Mean Reward")
plt.legend(loc="upper left")
plt.show()
```

```python
unity_walltime = unity_ml_agents["Wall time"] - unity_ml_agents["Wall time"][0]
godot_rllib_walltime = godot_rllib["Wall time"] - godot_rllib["Wall time"][0]
godot_sf_walltime = godot_sf["Wall time"] - godot_sf["Wall time"][0]
godot_crl_walltime = godot_crl["Wall time"] - godot_crl["Wall time"][0]
godot_sb3_walltime = godot_sb3["Wall time"] - godot_sb3["Wall time"][0]

plt.plot(unity_ml_agents.Step, unity_walltime / 60, label="Unity ML-Agents")
plt.plot(godot_rllib.Step, godot_rllib_walltime / 60, label="Godot RL Agents [RLlib]")
plt.plot(godot_sf.Step, godot_sf_walltime / 60, label="Godot RL Agents [Sample Factory]")
plt.plot(godot_crl.Step, godot_crl_walltime / 60, label="Godot RL Agents [CleanRL]")
plt.plot(godot_sb3.Step, godot_sb3_walltime / 60, label="Godot RL Agents [Stable Baselines 3]")
plt.xlabel("Steps")
plt.ylabel("Wall Time (in minutes)")
plt.legend(loc="upper left")
plt.show()
```
