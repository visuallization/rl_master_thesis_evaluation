```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

unity_ml_agents = pd.read_csv("./logs/unity_ml_agents.csv")
godot_rllib = pd.read_csv("./logs/godot_ray_rllib.csv")
godot_sf = pd.read_csv("./logs/godot_sample_factory.csv")
godot_sb3 = pd.read_csv("./logs/godot_stable_baselines_3.csv")
godot_crl = pd.read_csv("./logs/godot_clean_rl.csv")
```
